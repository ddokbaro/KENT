## 서지 정보 추출


```python
# 서지 정보 추출을 위한 반복문

level2 = 테스트전체데이터.find_all("level2")

서지정보 = []

for loop in level2:
  level2id = loop.get("id")
  mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()
  publisher =  loop.select_one("publication > publisher").get_text()
  dateIssued = loop.select_one("publication > dateIssued").get("date")
  종합 = [level2id, mainTitle, publisher, dateIssued ]
  서지정보.append(종합)


#서지정보
```

```python
# 추출한 서지 정보를 dataframe으로 전환
# 12건

df서지정보데이터 = pd.DataFrame(서지정보, columns=["level2id", "mainTitle", "publisher", "dateIssued"])
df서지정보데이터
```

## 본문 텍스트 추출

```python
# 텍스트 정보 추출을 위한 반복문

level3 = 테스트전체데이터.find_all("level3")

level3통합정보 = []

for loop in level3:
  level3id = loop.get("id")
  mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()
  subjectClass = loop.select_one("front > biblioData > subjectClass").get_text()
  pTitle = loop.select_one("text > content > paragraph > pTitle").get_text()

  paragraph = loop.find_all("paragraph")

  for paras in paragraph:
    para = paras.get_text()
    종합 = [level3id, mainTitle, subjectClass, pTitle, para]
    level3통합정보.append(종합)

# level3통합정보
```

```python
# 추출한 텍스트 정보를 dataframe으로 전환
# 3666건

df텍스트데이터 = pd.DataFrame(level3통합정보, columns=["level3id", "mainTitle", "subjectClass", "pTitle", "paragraph"])
df텍스트데이터
```

## 서지 정보 + 본문텍스트 병합

```python
# 서지정보와 본문텍스트를 병함(merge) 하기 위한 사전 작업
# df텍스트데이터의 level3id를 정규표현식을 통해 level2id 형식의 아이디로 변환

# ma_001_001000	    <--- level2의 ID
# ma_001_0010_0010	<--- level3의 ID


테스트_정규표현식일부추출 = df텍스트데이터.level3id.str.extract(r'(.{11})')
테스트_정규표현식일부추출
```

```python
# 서지정보와 본문텍스트를 병함(merge) 하기 위한 사전 작업
# df텍스트데이터의 level3id를 정규표현식을 통해 level2id 형식의 아이디로 변환하여 level2id 컬럼에 넣음

# ma_001	    <--- level1의 ID
# ma_001_001000	    <--- level2의 ID
# ma_001_0010_0010	<--- level3의 ID

df텍스트데이터['level2id'] = df텍스트데이터.level3id.str.extract(r'(.{11})')
df텍스트데이터
```

```python
# level1id를 중심으로 통합(merge) 수행
level2_level3통합 = pd.merge(df텍스트데이터, df서지정보데이터, on="level2id", how="left")
level2_level3통합
```

## 병합: 잡지정보 + 서지정보 + 본문텍스트

```python
# 잡지 정보 추출을 위한 반복문

level1 = 테스트전체데이터.find_all("level1")

잡지정보 = []

for loop in level1:
  level1id = loop.get("id")
  mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()
  begin = loop.select_one("front > biblioData > binderPeriod").get("begin")
  end = loop.select_one("front > biblioData > binderPeriod").get("end")
  종합 = [level1id, mainTitle, begin, end ]
  잡지정보.append(종합)

잡지정보
```

```python
# 추출한 잡지 정보를 dataframe으로 전환
# 1건

df잡지정보데이터 = pd.DataFrame(잡지정보, columns=["level1id", "mainTitle", "begin", "end"])
df잡지정보데이터
```

```python
# 서지정보와 본문텍스트를 병함(merge) 하기 위한 사전 작업
# df텍스트데이터의 level3id를 정규표현식을 통해 level1id 형식의 아이디로 변환하여 level1id 컬럼에 넣음

level2_level3통합['level1id'] = level2_level3통합.level3id.str.extract(r'(.{6})')
level2_level3통합
```

```python
# 잡지정보, 서지정보 본문텍스트를 병함(merge)

All_level통합 = pd.merge(level2_level3통합, df잡지정보데이터, on="level1id", how="left")
All_level통합 = All_level통합.replace('', np.nan) #빈 문자열을 NaN로 변환
All_level통합
```

## 서지 정보 추출

```python
# 서지 정보 추출을 위한 반복문
# 약 90초 소요

Path = "/content/KoreanModernMagazine/한국근현대잡지자료_20240327/"

서지정보  = []

for (root, directories, files) in os.walk(Path):
    for file in files:
        if '.xml' in file: # xml 확장자만 가져오기
            file_path = os.path.join(root, file) # 해당 줄까지 다수 파일 path 가져오기
            전체데이터 = BeautifulSoup(open(file_path), features="xml")

            level2 = 전체데이터.find_all("level2") ## 아래부터는 간략 서지정보 추출과 동일

            for loop in level2:
              level2id = loop.get("id")
              mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()
              publisher =  loop.select_one("publication > publisher").get_text()
              dateIssued = loop.select_one("publication > dateIssued").get("date")
              종합 = [level2id, mainTitle, publisher, dateIssued ]
              서지정보.append(종합)

#서지정보
```

```python
# 추출한 서지 정보를 dataframe으로 전환
# 2558건

df서지정보데이터 = pd.DataFrame(서지정보, columns=["level2id", "mainTitle", "publisher", "dateIssued"])
df서지정보데이터
```

## 본문 텍스트 추출

```python
# 서지 정보 추출을 위한 반복문
# 약 120초 소요

Path = "/content/KoreanModernMagazine/한국근현대잡지자료_20240327/"

level3통합정보  = []

for (root, directories, files) in os.walk(Path):
    for file in files:
        if '.xml' in file: # xml 확장자만 가져오기
            file_path = os.path.join(root, file) # 해당 줄까지 다수 파일 path 가져오기
            전체데이터 = BeautifulSoup(open(file_path), features="xml")

            level3 = 전체데이터.find_all("level3") ## 아래부터는 간략 본문 정보 추출과 동일

            for loop in level3:
              level3id = loop.get("id")
              mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()

              subjectClass = loop.select_one("front > biblioData > subjectClass") # subjectClass 데이터에 null 값이 있기에 체크 구문 추가
              if subjectClass is not None:
                subjectClass = subjectClass.get_text()
              else:
                subjectClass = ""

              pTitle = loop.select_one("text > content > paragraph > pTitle") # pTitle 데이터에 null 값이 있기에 체크 구문 추가
              if pTitle is not None:
                pTitle = pTitle.get_text()
              else:
                pTitle = ""


              paragraph = loop.find_all("paragraph")

              for paras in paragraph:
                para = paras.get_text()
                종합 = [level3id, mainTitle, subjectClass, pTitle, para]
                level3통합정보.append(종합)

#level3통합정보
```

```python
# 추출한 텍스트 정보를 dataframe으로 전환
# 588574건

df텍스트데이터 = pd.DataFrame(level3통합정보, columns=["level3id", "mainTitle", "subjectClass", "pTitle", "paragraph"])
df텍스트데이터
```

## 잡지 정보 추출

```python
# 잡지 정보 추출을 위한 반복문
# 약 90초 소요

Path = "/content/KoreanModernMagazine/한국근현대잡지자료_20240327/"

잡지정보  = []

for (root, directories, files) in os.walk(Path):
    for file in files:
        if '.xml' in file: # xml 확장자만 가져오기
            file_path = os.path.join(root, file) # 해당 줄까지 다수 파일 path 가져오기
            전체데이터 = BeautifulSoup(open(file_path), features="xml")

            level1 = 테스트전체데이터.find_all("level1") ## 아래부터는 간략 잡지 정보 추출과 동일

            for loop in level1:
              level1id = loop.get("id")
              mainTitle  = loop.select_one("front > biblioData > title > mainTitle").get_text()
              begin = loop.select_one("front > biblioData > binderPeriod").get("begin")
              end = loop.select_one("front > biblioData > binderPeriod").get("end")

              종합 = [level1id, mainTitle, begin, end ]
              잡지정보.append(종합)

#잡지정보
```

```python
# 추출한 잡지 정보를 dataframe으로 전환
# 84건

df잡지정보데이터 = pd.DataFrame(잡지정보, columns=["level1id", "mainTitle", "begin", "end"])
df잡지정보데이터
```

## 병합: 잡지정보 + 서지정보 + 본문텍스트

```python
# 서지정보와 본문텍스트를 병함(merge) 하기 위한 사전 작업
# df텍스트데이터의 level3id를 정규표현식을 통해 level2id 형식의 아이디로 변환하여 level2id 컬럼에 넣음

# ma_001	    <--- level1의 ID
# ma_001_001000	    <--- level2의 ID
# ma_001_0010_0010	<--- level3의 ID

df텍스트데이터['level2id'] = df텍스트데이터.level3id.str.extract(r'(.{11})')
#df텍스트데이터
```

```python
# level1id를 중심으로 통합(merge) 수행
level2_level3통합 = pd.merge(df텍스트데이터, df서지정보데이터, on="level2id", how="left")
#level2_level3통합
```

```python
# 서지정보와 본문텍스트를 병함(merge) 하기 위한 사전 작업
# df텍스트데이터의 level3id를 정규표현식을 통해 level1id 형식의 아이디로 변환하여 level1id 컬럼에 넣음

# ma_001	    <--- level1의 ID
# ma_001_001000	    <--- level2의 ID
# ma_001_0010_0010	<--- level3의 ID

level2_level3통합['level1id'] = level2_level3통합.level3id.str.extract(r'(.{6})')
#level2_level3통합
```

```python
# 잡지정보, 서지정보 본문텍스트를 병함(merge)
# 892852건

All_level통합 = pd.merge(level2_level3통합, df잡지정보데이터, on="level1id", how="left")
All_level통합 = All_level통합.replace('', np.nan) #빈 문자열을 NaN로 변환
All_level통합
```
